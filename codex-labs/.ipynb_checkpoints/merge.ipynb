```python
import pandas as pd
```

```python
df_students = pd.DataFrame({
    'Student_ID': [1,2,3],
    'Name': ['Alice', 'Bob', 'Charlie']
    })

print(df_students)
```

```python
df_grades = pd.DataFrame({
    'Student_ID': [1,2,3],
    'Grade': [90, 85, 88]
    })

print(df_grades)
```

```python
df_merged = pd.merge(df_students, df_grades, on='Student_ID', how='inner')
print(df_merged)
```
