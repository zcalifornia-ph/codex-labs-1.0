```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'ID': [1,2],
    'Score': [85, 90]
})

print(df1)
```

```python
df2 = pd.DataFrame({
    'ID': [3, 4],
    'Score': [88, 92]
})

print(df2)
```

```python
df_concat = pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
print(df_concat)
```
